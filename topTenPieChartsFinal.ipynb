```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.patches import ConnectionPatch
from itertools import product
import numpy as np
```

```python
# data
data_org = {
    'group': ['Eurasia', 'North America', 'Asia', 'Europe', 'Other', 'Unknown'],
    'freq.all': [36, 15, 4, 7, 2, 4],
    'freq.top10': [5, 3, 1, 1, 0, 0]
}

data_taxa = {
    'group': ['Fish', 'Plants', 'Mollusks', 'Algae', 'Crustaceans', 'Other'],
    'freq.all': [10, 32, 6, 4, 6, 10],
    'freq.top10': [5, 3, 2, 0, 0, 0]
}

data_vec = {
    'group': ['Ballast', 'Canal', 'Deliberate Release', 'Direct Accidental\n Release', 'Hitchhiker', 'Unknown'],
    'freq.all': [13, 1, 22, 15, 8, 9],
    'freq.top10': [4, 3, 1, 1, 1, 0]
}

# Sample data (replace this with your actual data)
patterns = ['', '.', '', '/', '', 'X']
colors = ['0.2', 'white', '0.5', 'white', '0.9', 'white']
category = ['(A) Continent\n of Origin', '(B) Taxonomic\n Group', '(C) Vector of\n Introduction']
```

```python
# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))

# PIE CHART 1
# Plot the first pie chart (freq.all)
nonzero_data_all = [(label, val) for label, val in zip(data_org['group'], data_org['freq.all']) if val != 0]
labels_all, values_all = zip(*nonzero_data_all)
wedges_all, autotexts_all = axes[0,0].pie(values_all, labels=values_all,
                                        colors=colors, startangle=90,
                                        radius = 1.2,
                                        pctdistance=0.1,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))
# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_all, patterns):
    wedge.set_hatch(hatch)
    
# Add title to figure    
axes[0,0].set_title('All Invasive Species', size=24, weight='bold', pad=30)

# Assuming axes is a list of axes and you want to set the label on the first axis
axes[0,0].set_ylabel(category[0], rotation=0, size=16, ha='left', weight='bold', labelpad=160)

# PIE CHART 2
# Plot the second pie chart (freq.top10)
nonzero_data_top10 = [(label, val) for label, val in zip(data_org['group'], data_org['freq.top10']) if val != 0]
labels_top10, values_top10 = zip(*nonzero_data_top10)
wedges_top10, autotexts_top10 = axes[0,1].pie(values_top10, labels=values_top10,
                                        colors=colors, startangle=90,
                                        radius=1.2,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))

axes[0,1].set_title('Top 10 Invasive Species', size=24, weight='bold', pad=30)

# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_top10, patterns):
    wedge.set_hatch(hatch)

# Legend for the first row
legend_org = fig.legend(labels_all, loc='center left', bbox_to_anchor=(0.9, 0.78), handlelength=2.5, handleheight=3, frameon=False, borderaxespad=4)

# Make legend text bold
for text in legend_org.get_texts():
    text.set_fontweight('bold')
    text.set_fontsize(14)
    

# PIE CHART 3
# Plot the first pie chart (freq.all)
nonzero_data_all = [(label, val) for label, val in zip(data_taxa['group'], data_taxa['freq.all']) if val != 0]
labels_all, values_all = zip(*nonzero_data_all)
wedges_all, autotexts_all = axes[1,0].pie(values_all, labels=values_all,
                                        colors=colors, startangle=90,
                                        pctdistance=0.85,
                                        radius = 1.2,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))
# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_all, patterns):
    wedge.set_hatch(hatch)

# Assuming axes is a list of axes and you want to set the label on the first axis
axes[1,0].set_ylabel(category[1], rotation=0, size=16, ha='left', weight='bold', labelpad=160)


# PIE CHART 4
# Plot the second pie chart (freq.top10)
nonzero_data_top10 = [(label, val) for label, val in zip(data_taxa['group'], data_taxa['freq.top10']) if val != 0]
labels_top10, values_top10 = zip(*nonzero_data_top10)
wedges_top10, autotexts_top10 = axes[1,1].pie(values_top10, labels=values_top10,
                                        colors=colors, startangle=90,
                                        radius=1.2,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))

# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_top10, patterns):
    wedge.set_hatch(hatch)
    

# Legend for the first row
legend_taxa = fig.legend(labels_all, loc='center left', bbox_to_anchor=(0.9, 0.47), handlelength=2.5, handleheight=3, frameon=False, borderaxespad=4)

# Make legend text bold
for text in legend_taxa.get_texts():
    text.set_fontweight('bold')
    text.set_fontsize(14)

# PIE CHART 5
# Plot the first pie chart (freq.all)
nonzero_data_all = [(label, val) for label, val in zip(data_vec['group'], data_vec['freq.all']) if val != 0]
labels_all, values_all = zip(*nonzero_data_all)
wedges_all, autotexts_all = axes[2,0].pie(values_all, labels=values_all,
                                        colors=colors, startangle=90,
                                        pctdistance=0.85,
                                        radius=1.2,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))
# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_all, patterns):
    wedge.set_hatch(hatch)

# Assuming axes is a list of axes and you want to set the label on the first axis
axes[2,0].set_ylabel(category[2], rotation=0, size=16, ha='left', weight='bold', labelpad=160)

# PIE CHART 6
# Plot the second pie chart (freq.top10)
nonzero_data_top10 = [(label, val) for label, val in zip(data_vec['group'], data_vec['freq.top10']) if val != 0]
labels_top10, values_top10 = zip(*nonzero_data_top10)
wedges_top10, autotexts_top10 = axes[2,1].pie(values_top10, labels=values_top10,
                                        colors=colors, startangle=90,
                                        radius=1.2,
                                        textprops=dict(color="black", size=12),
                                        wedgeprops=dict(width=1.2, edgecolor='black', linewidth=1.5))

# Apply hatch patterns to wedges
for wedge, hatch in zip(wedges_top10, patterns):
    wedge.set_hatch(hatch)


# Legend for the first row
legend_vec = fig.legend(labels_all, loc='center left', bbox_to_anchor=(0.9, 0.16), handlelength=2.5, handleheight=3, frameon=False, borderaxespad=4)

# Make legend text bold
for text in legend_vec.get_texts():
    text.set_fontweight('bold')
    text.set_fontsize(14)

# Adjust layout for better spacing
plt.tight_layout()

# save fig
plt.savefig('test.jpg', dpi = 300, bbox_inches='tight') 
```
